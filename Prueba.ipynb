```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import dataloader
```

```python
data_dir = "/home/quimisagi/Daigaku/Datasets/adobe240/test"
dataset = dataloader.SuperSlowMoDataset(data_dir)
```

```python
def display_triplets(dataset, num_triplets=5):
    # Ensure the number of triplets does not exceed the available data
    total_frames = len(dataset)
    max_triplets = total_frames // 3  # Each triplet consists of 3 frames
    
    for i in range(num_triplets):
        # Ensure we don't go out of bounds
        if i >= max_triplets:
            break
        
        # Calculate the index for the triplet
        index = i * 3
        
        # Get the paths for the triplet
        sample_paths = dataset[index:index + 3]  # Extract 3 paths for the triplet
        
        # Display the triplet
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        
        for j, frame_path in enumerate(sample_paths):
            # Load the image from the path
            image = Image.open(frame_path)
            axes[j].imshow(image)
            axes[j].axis('off')
            axes[j].set_title(f'Frame {j + 1}')
        
        plt.show()

# Call the function with your dataset
display_triplets(dataset, num_triplets=5)
```
